<b> A diagnostic test has a 98% probability of giving a positive result when applied to a person suffering from Thripshaw's Disease, and 10% probability of giving a (false) positive when applied to a non-sufferer. It is estimated that 0.5 % of the population are sufferers. Suppose that the test is now administered to a person whose disease status is unknown. Calculate the probability that the test will:

1. Be positive
2. Correctly diagnose a sufferer of Thripshaw's
3. Correctly identify a non-sufferer of Thripshaw's
4. Misclassify the person
</b>

First thing need to do is draw a Tree...



<b>1.  Be Positive <b>

P(Infected|Positive) = (0.98)(0.005)/[(0.005*0.98)+(0.98*0.1)] = 0.0469

P(Positive) = (0.98*0.005)/0.0469 = 0.1044

<b>2.  Correctly Diagnose a suffer of Thripshaw <b>

P(Positive | Infected) = 0.98

<b>3.  Correctly Diagnose a Non suffer of Thripshaw <b>

P(Negative | Not Infected) = 0.9

<b>4.  Misclassify <b>

First, find the probability will classify correctly:
P(Not Infected)*P(Negative | Not Infected) + P(Infected)*P(Positive|Infected)
(0.995*0.9) + (0.005*0.98) = 0.9004

Therefore the probability of Misclassifying is 1 - 0.9004 = 0.0996
