```python
%time
stuff = [random.randrange(1000) for _ in range(1000)]

def bubble_sort(array):
    swapped = True
    while swapped:
        swapped = False
        for idx in range(len(array) - 1):
            if array[idx] > array[idx+1]:
                array[idx],array[idx+1] = array[idx+1],array[idx]
                swapped = True
    return array
bubble = bubble_sort(stuff)
```
